# Life Expectancy and GDP

#### Introduction

The purpose of this project is to analyze data on GDP and life expectancy from the World Health Organization and the World Bank to try and identify the relationship between the GDP and life expectancy of six countries.

Some questions to answer using our data:

* Has life expectancy increased over time in the six nations?
* Has GDP increased over time in the six nations?
* Is there a correlation between GDP and life expectancy of a country?
* What is the average life expectancy in these nations?
* What is the distribution of that life expectancy?

#### Import Modules

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

#### Import and Inspect the data

```python
#inspect the data
df = pd.read_csv('all_data.csv')
print(df.head())
df.info()
```

```python
print(df.Country.unique())
print(df.Year.unique())
```

Above we can see the data set includes 6 different countries and data from 2000-2015

```python
#Some Data Tyding
df = df.rename({'Life expectancy at birth (years)' : 'LEyrs'}, axis='columns')
```

<B>Life Expectancy<B>

```python
#Univariate Analysis - Life Expectancy

#boxplot of life expectancy
plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
sns.boxplot(y = df['LEyrs'])
plt.ylabel('Life Expectancy at birth (years)')
plt.title('Distribution of Life Expectancy (years)')

#histogram of life expectancy
plt.subplot(1,2,2)
sns.histplot(df['LEyrs'])
plt.xlabel('Life Expectancy at birth (years)')
plt.title('Distribution of Life Expectancy (years)')

plt.show()
plt.clf()

print('The mean Life expectancy at birth is ' + str(np.mean(df['LEyrs'])) + ' years')
print('The median Life expectancy at birth is ' + str(np.median(df['LEyrs'])) + ' years')
```

A first look shows that life expectancy is left-skewed with a mean of 72.7896 years and median of 76.75 years.

<B>GDP<B>

```python
#Univariate Analysis - GDP

#boxplot of GDP
plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
sns.boxplot(y = df['GDP'])
plt.ylabel('GDP in trillions of dollars')
plt.title('Distribution of GDP')

#histogram of GDPb with outliers removed
plt.subplot(1,2,2)
sns.histplot(df['GDP'])
plt.title('Distribution of GDP')
plt.xlabel('GDP in trillions of dollars')
plt.show()
plt.clf()

print('The mean GDP is ' + str(np.mean(df['GDP'])) )
print('The median GDP is ' + str(np.median(df['GDP'])))
```

A first look at GDP shows the data is right skewed with a mean of 3.88 trillion dollars and median of 1.28 trillion dollars.

<B>Averages<B>

```python
dfMeans = df.drop("Year", axis = 1).groupby("Country").mean().reset_index()
dfMeans
```

```python
plt.figure(figsize= (15,8))
plt.subplot(1,2,1)
sns.barplot(x="Country", y="LEyrs", data=dfMeans)
plt.ylabel("Life expectancy at birth (years)")
plt.xticks(rotation=45)
plt.title('Mean of Life Expectancy')

plt.subplot(1,2,2)
sns.barplot(x="Country", y="GDP", data=dfMeans)
plt.ylabel("GDP (trillions)");
plt.xticks(rotation=45)
plt.title('Mean of GDP')

plt.show()
plt.clf()
```

The first plot on the left shows that most countries have average life expectancies somewhere in the mid 70s, with the exception of Zimbabwe, which is closer to 50.

The second plot shows a large disparity in mean GDP figures. The US has much larger GDP than any other countries. 

<B> Life Expectancy by Country <B>

```python
#Bivariate Analysis

#sns.boxplot(data = df, y = 'Country', x = 'LEyrs')
plt.figure(figsize=(15,8)),
plt.subplot(1,2,1)
sns.violinplot(data = df, x = 'Country', y = 'LEyrs')
plt.xticks(rotation=45)
plt.title('Distribution of Life Expectancy by Country')

plt.subplot(1,2,2)
sns.histplot(data = df, x = 'LEyrs', hue = 'Country', alpha = 0.5)
plt.title('Distribution of Life Expectancy by Country')
plt.show()
plt.clf()
```

The above plots look into Life Expectancy in more detail and break it down by conntry. We can see from the violin plots, that most country's life expectancy has a short range with the exception of Zimbabwe which has a larger range between roughly 35-78 years.

<B> GDP by Country <B>

```python
plt.figure(figsize=(15,8)),
plt.subplot(1,2,1)
sns.boxplot(data = df, x = 'Country', y = 'GDP')
plt.ylabel('GDP in trillions of dollars')
plt.xticks(rotation=45)
plt.title('Distribution of GDP by Country')

plt.subplot(1,2,2)
sns.histplot(data = df, x = 'GDP', hue = 'Country', alpha = 0.5)
plt.xlabel('GDP in trillions of dollars')
plt.title('Distribution of GDP by Country')
plt.show()
plt.clf()
```

Above we look at the distribution of GDP broken down by each country. We can see that China has the largest range of GDP of all contries.

<B> Life Expectancy & GDP by Country over time <B>

```python
plt.figure(figsize=(15,8)),
plt.subplot(1,2,1)
sns.lineplot(data = df, x = 'Year', y = 'LEyrs', hue = 'Country')
plt.legend(loc = 'lower right')
plt.title('Life Expectancy over time')
plt.subplot(1,2,2)
sns.lineplot(data = df, x = 'Year', y = 'GDP', hue = 'Country')
plt.legend(loc = 'upper left')
plt.title('GDP (trillions) over time')
plt.show()
plt.clf()
```

Above we look at both Life Expectancy and GDP over timne for each country. We can spot that both seem to increase over time for all countries but it can be difficult to sport for some countries because they're all plotted on the same axis. Below I have plotted each country on it's own axis to give us a better look.

```python
CountryGDP = sns.FacetGrid(data=df, col = 'Country', hue = 'Country', col_wrap = 3, sharey = False, sharex = False)
CountryGDP.map(sns.lineplot, 'Year', 'GDP').set_axis_labels('Year','GDP')
```

From the above plots we can confirm that GDP for all counries has infact increased over time. China has seen a fairly steady upward trend overtime but it does seem to increase at a faster pace in the later years. Zimbabwe begins with a decrease in GDP before a very sharp and rapid increase in the later years. The other countries seem to be a little more volatile in their trend but still an overall upward trend.

```python
CountryGDP = sns.FacetGrid(data=df, col = 'Country', hue = 'Country', col_wrap = 3, sharey = False, sharex = False)
CountryGDP.map(sns.lineplot, 'Year', 'LEyrs').set_axis_labels('Year','Life expectancy at birth(years)')
```

We can see more clearly from the above graphs that life expectancy has increased for each country overtime. We see Zimbabwe's more dramatic increase from their graph

<B>Relationship of Life Expectancy & GDP <B>

```python
CountryGDP = sns.FacetGrid(data=df, col = 'Country', hue = 'Country', col_wrap = 3, sharey = False, sharex = False)
CountryGDP.map(sns.scatterplot , 'LEyrs', 'GDP').set_axis_labels("Life expectancy at birth (years)", "GDP in Trillions of U.S. Dollars")
```

Now we look at the correlation between Life Expectancy and GDP. Its very easy to see from the abpve graphs that their is a clear positive correlation between to two variables.
